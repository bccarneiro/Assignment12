```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
```

```python
sqladb = pd.read_csv('adult.csv')
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

Select 10 records

```python
pysqldf('SELECT * FROM sqladb LIMIT 10')
```

Average hours per week of all men who are working in private sector

```python
pysqldf('''SELECT AVG("hours-per-week") FROM sqladb WHERE gender='Male' AND workclass='Private';''')
```

Frequency table for education, occupation and relationship separately

```python
pysqldf('''SELECT education, COUNT(education) FROM sqladb GROUP BY education;''')
```

```python
pysqldf('''SELECT occupation, COUNT(occupation) FROM sqladb GROUP BY occupation;''')
```

```python
pysqldf('''SELECT relationship, COUNT(relationship) FROM sqladb GROUP BY relationship;''')
```

Are there any people who are married, working in private sector and having a masters degree

```python
pysqldf('''SELECT * FROM sqladb WHERE "marital-status"='Married-civ-spouse' AND workclass='Private' AND education='Masters';''')
```

What is the average, minimum and maximum age group for people working in different sectors

```python
pysqldf('''SELECT workclass, AVG(age), MIN(age), MAX(age)  FROM sqladb GROUP BY workclass;''')
```

Age distribution by country

```python
pysqldf('''
SELECT 
    "native-country", 
    SUM(CASE WHEN age < 18 THEN 1 ELSE 0 END) AS [Under 18], 
    SUM(CASE WHEN age BETWEEN 18 AND 65 THEN 1 ELSE 0 END) AS [18-65], 
    SUM(CASE WHEN age > 65 THEN 1 ELSE 0 END) AS [Above 65]  
FROM sqladb 
GROUP BY "native-country";
''')
```

Compute a new column as 'Net-Capital-Gain'from the two columns 'capital-gain' and 'capital-loss'

```python
pysqldf('''
SELECT 
    "capital-gain"-"capital-loss"  
FROM sqladb;
''')
```
